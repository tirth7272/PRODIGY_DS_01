```python
countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']


filtered_data = data[data['Country Name'].isin(countries_to_plot)]


plt.bar (filtered_data['Country Name'], filtered_data['2022'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2022')
plt.title('Population of selected countries in 2022')
plt.xticks(rotation=45, ha='right') 
plt.show()
```

```python
df_melted = data.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
                      var_name='Year',
                      value_name='Population')
```

```python
df_melted['Year'] = pd.to_numeric(df_melted['Year'], errors='coerce')
df_melted['Population'] = pd.to_numeric(df_melted['Population'], errors='coerce')
```

```python
df_clean = df_melted.dropna(subset=['Population', 'Year'])
```

```python
selected_countries = ['China', 'India', 'United States', 'Japan', 'South Korea', 'Canada']
df_trend = df_clean[df_clean['Country Name'].isin(selected_countries)]
```

```python
fig, axs = plt.subplots(len(selected_countries), 1, figsize=(10, 20), sharex=True)

for i, country in enumerate(selected_countries):
    df_country = df_trend[df_trend['Country Name'] == country]
    axs[i].plot(df_country['Year'], df_country['Population'], marker='o', color='tab:blue')
    axs[i].set_title(f'Population Trend for {country}')
    axs[i].set_yscale('log')
    axs[i].set_ylabel('Population')

plt.xlabel('Year')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt  # In case you haven't imported matplotlib yet

pivot_table = df_clean.pivot_table(index='Country Name', columns='Year', values='Population')

plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table.corr(), cmap='coolwarm', linecolor='white', linewidths=0.1)
plt.title('Heatmap of Yearly Population Correlation')
plt.xlabel('Year')
plt.ylabel('Year')
plt.show()
```
